```python
import pandas as pd
import matplotlib.pyplot as plt
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
moviedata= pd.read_csv('IMDB Dataset.csv')
```

```python
# Remove rows with missing values (NaN) from the moviedata DataFrame.
moviedata.dropna(inplace=True)
```

```python
moviedata
```

```python
def remove_unwanted(review):
    """
    This function cleans and preprocesses a movie review by:
        1. Removing non-alphabetic characters.
        2. Converting to lowercase.
        3. Removing English stopwords.

    Args:
        review (str): The raw movie review text.

    Returns:
        str: The cleaned and preprocessed review text.
    """
    review = re.sub("[^a-zA-Z]", " ", review)  # Remove non-alphabetic characters, replacing with space
    words = review.lower().split()  # Convert to lowercase and split into words
    filtered_words = [word for word in words if word not in stopwords.words('english')]  # Filter out stopwords
    return " ".join(filtered_words)  # Join the remaining words back into a string
```

```python
# Apply the function to each review in the DataFrame
moviedata['review'] = moviedata['review'].apply(remove_unwanted)  
```

```python
moviedata
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer  # Import the TF-IDF vectorizer class

# Create a TF-IDF vectorizer with a maximum of 5000 features
tfid_vectorizer = TfidfVectorizer(max_features=5000)
```

```python
# Transform the movie reviews into TF-IDF feature vectors
BinaryMovieReviews = tfid_vectorizer.fit_transform(moviedata['review']).toarray() 
BinaryMovieReviews
```

```python
# Replace textual sentiment labels ('positive', 'negative') with numerical labels (0, 1) in the 'sentiment' column
moviedata['sentiment'] = moviedata['sentiment'].replace(['positive','negative'],[0,1])
moviedata
```

```python
BinaryMovieReviews
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Initialize a logistic regression model
LRmodel = LogisticRegression()
```

```python
# Split the data into training and testing sets
reviews_train,reviews_test,sentiment_train,sentiment_test = train_test_split(BinaryMovieReviews,moviedata['sentiment'],test_size=0.2)
```

```python
# Fit the logistic regression model on the training data. 
LRmodel.fit(reviews_train,sentiment_train)
```

```python
# Make predictions on the testing set using the trained logistic regression model. 
predict_test = LRmodel.predict(reviews_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Create a confusion matrix comparing the true labels (`sentiment_test`) with the predicted labels (`predict_test`).
cm = confusion_matrix(sentiment_test, predict_test)
# Generate a visual representation of the confusion matrix
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,  # The confusion matrix data
    display_labels=LRmodel.classes_  # The class labels (e.g., 'Negative', 'Positive')
)
# Display the confusion matrix plot
disp.plot()
plt.show() 
```

```python
# Predict the sentiment of the sentence using the trained model
sentence="the movie is really good"
str= ' '.join(word for word in sentence.split() if word.lower() not in stopwords.words('english'))
sampletext=tfid_vectorizer.transform([str]).toarray()
predict=LRmodel.predict(sampletext)
predict # '0' response for the positive
```

```python
import pickle

# Save the model
with open('LRmodel.pkl', 'wb') as file:
    pickle.dump(LRmodel, file)
with open('tfid_vectorizer.pkl','wb') as file:
    pickle.dump(tfid_vectorizer,file)
```
